```python
#determinant of a matrix
A1 = np.array([[1,2],[3,4]])
det_a = np.linalg.det(A1)
print (det_a, ": is its determininat")

#Rank of a matrix (number of linearly independent row or column)
A2= np.array([[1,2,3,4],
              [2,4,6,7], 
              [9,5,6,7],
              [5,6,7,1]])

rank = np.linalg.matrix_rank(A2)
print ("rank of A2 is: ", rank)

#eigenvalue and eigenvector

eigenvalue, eigenvector = np.linalg.eig(A2)

print("eigenvalues are:", eigenvalue)
print("eigenvectors are:", eigenvector)

# Specteral decomposition of a matix

A3 = np.array([[1,2,3],
               [2,1,5],
               [3,5,0]])
eigenvalue, eigenvector = np.linalg.eig(A3)

print("eigenvector of A3, symmertical matrix is: ", eigenvector)
```

```python
# Test: when you multiply a non symmetrical matrix with its transpose, you get a symmertical martix
A2= np.array([[1,2,3,4],
              [2,4,6,7], 
              [9,5,6,7],
              [5,6,7,1]])

A2_t = A2.T

sym_matrix = np.dot(A2, A2_t)

print("is it symmetrical?: ", sym_matrix)
```

```python
#Inverse of a matrix

product = np.dot(A2, np.linalg.inv(A2))

print ("identity matrix?: ", product)
```

```python
# Specteral Decomposition

A4 = np.array([[1,2,4],
               [2,1,3],
               [4,3,6]])

eigenvalue, eigenvector = np.linalg.eig(A4)

D = np.diag(eigenvalue)

A_decomposed = eigenvector @ D @ eigenvector.T

print ("Specteral decomposition: \n", eigenvector)

#each column in eigenvector matrix represent eigenvector associated with each eigen vector


# Singular value decomposition 

U, E, Vt = np.linalg.svd(A4)

print ("\n \nU is", U)
print ("\n E is\n", E)
print ("\n Vt is\n", Vt)


print(np.linalg.svd(A4))
```

```python
# Principle component analysis

from sklearn.decomposition import PCA

# Sample dataset
X = np.array([[1, 2, 4],
              [2, 1, 3],
              [4, 3, 6]])

# Initialize PCA with 2 components (dimensionality reduction to 2D)
pca = PCA(n_components=2)

# Fit PCA on the data
X_pca = pca.fit_transform(X)

# Print results
print("Principal Components:\n", pca.components_)
print("Explained Variance (Eigenvalues):\n", pca.explained_variance_)
print("Transformed Data (Projection onto Principal Components):\n", X_pca)
```
